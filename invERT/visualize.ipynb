```python
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt
import sys
```

```python

current_dir = Path.cwd()
result_folder = current_dir.parent / "results"
result_folder /= "batch_size_8_18-12-2024_10h54"
repetitions = 5
```

```python
arrays = np.load(result_folder / "loss_arrays.npz")
burn_in: int = 40
# Plot results
for repetition in range(repetitions):
    color = plt.cm.tab10(repetition % 10)
    plt.plot(
        arrays["loss_arrays"][repetition][burn_in:],
        label=f"Repetition {repetition + 1}",
        color=color,
    )
    plt.plot(
        arrays["test_loss_arrays"][repetition][burn_in:],
        label=f"Repetition {repetition + 1} (test)",
        color=color,
        alpha=0.7
    )
plt.xlabel("Iteration")
plt.ylabel("Loss")
plt.title("Training and testing loss")
plt.legend()
plt.show()
```
